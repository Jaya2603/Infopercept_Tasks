```python
# 61] Write a program to accept any number and count how many odd digits and how many even digits in that number.

number = input("Enter a number: ")

odd_count = 0
even_count = 0

for num in number:
    if int(num) % 2 != 0:
        odd_count += 1
    if int(num) % 2 == 0:
        even_count += 1

print("Odd digits:", odd_count)
print("Even digits:",even_count)
```

```python
# 62] Write a program, which will read an integer value for a base, then read a positive integer raised to that base and 
#print its value.

base = int(input("Enter the base: "))
exponent = int(input("Enter a positive integer raised to that base: "))

result = base ** exponent

print(f"{base} raised to the power of {exponent} is: {result}")
```

```python
# 63] Write a menu driven program which has the following options:
# a] Addition of 2 numbers
# b] Subtraction of 2 numbers
# c] Multiplication of 2 numbers
# d] Division of 2 numbers
# e] Exit.
# Make use of SWITCH statement.

def cal():
    print("\nCALCULATE:")
    print("a) Addition of 2 numbers")
    print("b) Subtraction of 2 numbers")
    print("c) Multiplication of 2 numbers")
    print("d) Division of 2 numbers")
    print("e) Exit")

while True:
    cal()
    choice = input("Choose an option: ")

    if choice == 'e':
        print("Exit")
        break

    elif choice in ['a', 'b', 'c', 'd']:
        num1 = float(input("Enter the first number: "))
        num2 = float(input("Enter the second number: "))

        if choice == 'a':
            result = num1 + num2
        elif choice == 'b':
            result = num1 - num2
        elif choice == 'c':
            result = num1 * num2
        elif choice == 'd':
            result = num1 / num2 if num2 != 0 else "Cannot divide by zero."

        print(f"Result: {result}")
    else:
        print("Invalid choice, please try again.")
```

```python
# 64] Write a menu driven program which has the following options:
# a] Positive or Negative.
# b] Even or Odd.
# c] Leap Year or Not Leap Year.
# d] Maximum of two numbers.
# e] Exit.
# Make use of SWITCH statement.

while True:
    print("\nCALCULATE:")
    print("a) Positive or Negative")
    print("b) Even or Odd")
    print("c) Leap Year or Not Leap Year")
    print("d) Maximum of two numbers")
    print("e) Exit")

    choice = input("Choose an option:")

    if choice == 'e':
        print("Exit")
        break

    elif choice == 'a':
        num = float(input("Enter a number: "))
        result = "Positive" if num >= 0 else "Negative"

    elif choice == 'b':
        num = int(input("Enter a number: "))
        result = "Even" if num % 2 == 0 else "Odd"

    elif choice == 'c':
        year = int(input("Enter a year: "))
        result = "Leap Year" if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0) else "Not a Leap Year"

    elif choice == 'd':
        num1 = int(input("Enter the first number: "))
        num2 = int(input("Enter the second number: "))
        result = f"Maximum of the two numbers is: {max(num1, num2)}"

    else:
        print("Invalid choice, please try again.")
        continue

    print(f"Result: {result}")
```

```python
# 65] Write a menu driven program which has the following options:
# a] Factorial of a number.
# b] Reverse of a number.
# c] Sum of Digits of a number.
# d] Count Digits of a number.
# e] Exit.
# Make use of SWITCH statement.

def cal():
    print("\nCALCULATE:")
    print("a) Factorial of a number")
    print("b) Reverse of a number")
    print("c) Sum of Digits of a number")
    print("d) Count Digits of a number")
    print("e) Exit")

while True:
    cal()
    choice = input("Choose an option: ").lower()

    if choice == 'e':
        print("Exit")
        break

    num = int(input("Enter a number: "))

    if choice == 'a':
        if num < 0:
            result = "Factorial is not defined for negative numbers."
        else:
            result = 1
            i = 1
            while i <= num:
                result *= i
                i += 1

    elif choice == 'b':
        result = int(str(num)[::-1])

    elif choice == 'c':
        result = sum(int(digit) for digit in str(num))

    elif choice == 'd':
        result = len(str(num))

    else:
        print("Invalid choice, please try again.")
        continue

    print(f"Result: {result}")
```
